# Sigmoid is a linear combination of Tanh

## What is sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-10,10,0.1)
sig_y = sigmoid(x)
plt.plot(x,sig_y)
plt.title("sigmoid",fontsize=20)
```

## What is tanh

```python
def tanh(x):
    return np.tanh(x)
```

```python
tanh_y = tanh(x)
plt.plot(x,tanh_y)
plt.title("tanh",fontsize=20)
```

## A linear combination of tanh

```python
def linear_tanh(x):
    return (1+tanh(x))/2
```

```python
linear_tanh_y = linear_tanh(x)
plt.plot(x,linear_tanh_y,label="linear_tanh")
plt.plot(x,sig_y,label="sigmoid")
plt.legend()
plt.title("comparision of sigmoid and linear combination of tanh",fontsize=20)
```
